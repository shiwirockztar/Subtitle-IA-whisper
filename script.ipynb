```
import whisper

ruta_video = "/content/drive/MyDrive/3012468.mp4"

model = whisper.load_model("medium")
result = model.transcribe(ruta_video)

# ---- GENERAR ARCHIVO SRT MANUALMENTE ----
def format_timestamp(seconds):
    millis = int((seconds % 1) * 1000)
    seconds = int(seconds)
    hrs = seconds // 3600
    mins = (seconds % 3600) // 60
    secs = seconds % 60
    return f"{hrs:02d}:{mins:02d}:{secs:02d},{millis:03d}"

with open("subtitulos.srt", "w", encoding="utf-8") as srt:
    for i, seg in enumerate(result["segments"]):
        start = format_timestamp(seg["start"])
        end = format_timestamp(seg["end"])
        text = seg["text"].strip()

        srt.write(f"{i+1}\n{start} --> {end}\n{text}\n\n")

# ---- ARCHIVO DE TEXTO SIMPLE ----
with open("subtitulos.txt", "w", encoding="utf-8") as f:
    f.write(result["text"])

print("Subt√≠tulos generados correctamente.")
```
